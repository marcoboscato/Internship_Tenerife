```python
# COMPARISON FUNCTION

def compare_columns(column1, column2, column_name):
    diff = np.setdiff1d(column1, column2)
    if len(diff) > 0:
        print(f'(my output) number different values in {column_name}: {len(diff)}')
    else:
        print(f'All values are the same in {column_name}.')

columns_legend = output.columns.values

print('My output file shape: ',output.shape)
print('Comparison output file shape: ',output0.shape)

for column in columns_legend:
    compare_columns(output[column].values, output0[column].values, column)
```
